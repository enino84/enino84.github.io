# **Duffing Equation**

Duffing Equation

\begin{eqnarray}
x''+\delta x' + \alpha x +\beta x^3= \gamma \cos(\omega t)
\end{eqnarray}

this is equivalent to

\begin{eqnarray}
x''=-\delta x' - \alpha x -\beta x^3+ \gamma \cos(\omega t)
\end{eqnarray}

https://numpy.org/

```python
import numpy as np
```

```python
#(alpha, beta, delta, gamma, omega)
def f(x, t, args=(1, 5, 0.02, 8, 0.65)):
  alpha = args[0];
  beta = args[1];
  delta = args[2];
  gamma = args[3];
  omega = args[4];
  return (x[1], -delta*x[1] - alpha*x[0] - beta * x[0]**3 + gamma * np.cos(omega * t))
```

```python
from scipy.integrate import odeint
```

Initial condition $x'(0) = 1$ and $x(0) = 0$, $t \in [0,\, 200]$, and the sample size equals 200

```python
x0 = np.array([1, 0], dtype=np.float32);
t = np.linspace(0,200, 1000);

x_samples = [];
samples = 200;
for i in range(0, samples):
  xi = x0 + 0.1*np.random.randn(2,); #perturbation ~ 0.1
  xs = odeint(f, xi, t);
  x_samples.append(xs);
```

```python
import matplotlib.pyplot as plt
```

In the beggining, small perturbations are not amplified

```python
plt.figure(figsize=(20,4))
for i in range(0, samples):
  x = x_samples[i];
  plt.plot(t[:500], x[:500,0], '-');

plt.xlabel('$t$', fontdict={'size':16});
plt.ylabel('$x$', fontdict={'size':16});
```

At the end, small perturbations have been amplified by the non-linear dynamics of the system

```python
plt.figure(figsize=(20,4))
for i in range(0, samples):
  x = x_samples[i];
  plt.plot(t[500:], x[500:,0], '-');

plt.xlabel('$t$', fontdict={'size':16});
plt.ylabel('$x$', fontdict={'size':16});
```

```python
plt.figure(figsize=(20,4))
for i in range(0, samples):
  x = x_samples[i];
  plt.plot(t, x[:,0], '-k', alpha=.5, lw=0.2);

plt.xlabel('$t$', fontdict={'size':16});
plt.ylabel('$x$', fontdict={'size':16});
```

Initial distribution (initial conditions)

\begin{eqnarray}
\begin{bmatrix}
x(0) \\
x'(0)
\end{bmatrix} + 
\begin{bmatrix}
\epsilon_0 \\
\epsilon_1
\end{bmatrix}
\end{eqnarray}

where $\epsilon_i \sim \mathcal{N}(0,\, 0.1^2)$

# **Lorenz 96**


The Lorenz96 model is a simplified model of atmospheric dynamics that was introduced by Edward Lorenz in 1996. The model consists of a set of coupled ordinary differential equations that describe the evolution of a one-dimensional array of variables representing atmospheric variables. The Lorenz96 model is known for its chaotic behavior, which makes it a useful testbed for studying the behavior of data assimilation algorithms.

The Lorenz96 model has been used extensively in the data assimilation community as a benchmark for testing and comparing different data assimilation algorithms. This is because the model is simple enough to allow for easy testing and analysis of assimilation methods, while also exhibiting the non-linear and chaotic behavior that is characteristic of more complex atmospheric models.

In particular, the Lorenz96 model has been used to study the behavior of ensemble-based data assimilation methods such as the Ensemble Kalman Filter (EnKF). By simulating the evolution of the Lorenz96 model, researchers can generate synthetic observations and test the ability of different assimilation methods to recover the true state of the system. This allows for the comparison of different algorithms and the identification of strengths and weaknesses in different approaches.

The Lorenz96 model has played a significant role in advancing the field of data assimilation by providing a simple and versatile testbed for the development and testing of new assimilation methods.

\begin{eqnarray}
\frac{dx_i}{dt} = (x_{i+1}-x_{i-2})x_{i-1} - x_i + F
\end{eqnarray}

When $F=8$, the model exhibit chaotic behavior

```python
def lorenz96(x, t, args=(8,)):
  n = x.size;
  F = args[0];
  return [(x[(i+1)%n]-x[i-2])*x[i-1]-x[i]+F for i in range(0,n)]; #list comprehension
```

```python
n = 40;
x0 = np.random.randn(n)
t = np.linspace(0,200, 5000);

xs = odeint(lorenz96, x0, t);
```

# **Example Linear Model**

Synthetic Data

```python
np.random.seed(10)

n = 100;
y0 = np.zeros((100,));
y1 = np.ones((100,));

X0 = np.array([0,0]) + np.random.randn(100,2)
X1 = np.array([3,3]) + np.random.randn(100,2)

plt.plot(X0[:,0], X0[:,1], 'bo');
plt.plot(X1[:,0], X1[:,1], 'ro');
```

Problem

```python
x = np.linspace(-3,6,50);
y = np.linspace(-3,6,50);

for xi in x:
  for yi in y:
    plt.plot(xi,yi,'.k')


plt.plot(X0[:,0], X0[:,1], 'bo');
plt.plot(X1[:,0], X1[:,1], 'ro');
```

# $y = \alpha_1 x_1 + \alpha_2 x_2 + \epsilon$

```python

y = np.hstack((y0,y1));
X = np.vstack((X0,X1));

alp = np.linalg.solve(X.T@X, X.T@y);
```

```python
y_est = X@alp;

plt.plot(y_est[:100], '-b')
plt.plot(y_est[100:], '-r')
```

```python
x = np.linspace(-3,6,50);
y = np.linspace(-3,6,50);

for xi in x:
  for yi in y:
    y_i = xi*alp[0]+yi*alp[1];
    if y_i>0.5:
      plt.plot(xi, yi, 'r.');
    else:
      plt.plot(xi, yi, 'b.');


plt.plot(X0[:,0], X0[:,1], 'bo');
plt.plot(X1[:,0], X1[:,1], 'ro');
```

# $y = \alpha_1 x_1^2 + \alpha_2 x_2^2 + \epsilon$

```python
y = np.hstack((y0,y1));
X = np.vstack((X0**2,X1**2));

alp = np.linalg.solve(X.T@X, X.T@y);
```

```python
y_est = X@alp;

plt.plot(y_est[:100], '-b')
plt.plot(y_est[100:], '-r')
```

```python
x = np.linspace(-3,6,50);
y = np.linspace(-3,6,50);

for xi in x:
  for yi in y:
    y_i = alp[0]*(xi**2)+alp[1]*(yi**2);
    if y_i>0.25:
      plt.plot(xi, yi, 'r.');
    else:
      plt.plot(xi, yi, 'b.');


plt.plot(X0[:,0], X0[:,1], 'bo');
plt.plot(X1[:,0], X1[:,1], 'ro');
```
